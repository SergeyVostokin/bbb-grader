**Шаг 2.** Скопируйте список присутсвующих в файл ***"список присутсвующих.txt"***

**Шаг 3.** Вывести *распознанные пары* вида ***"ФИО в списке группы" -- "ФИО в списе присутсвующих"***   

```python
stars = []
points = []
otsut = []

def checkPres(name):
    newName = name[5:].strip()
    if (name.__contains__("Сортировать по фамилии:")):
        return False
    if( not name[:4].isdigit()):
        otsut.append(name.strip())
        return False
    with open("список группы.txt", "r",encoding= 'UTF-8') as file:
        dd = file.readline()
        while(dd!=""):
            if(dd.__contains__(newName)):
                file.close()
                return True
            dd = file.readline()
        otsut.append(name.strip())
    file.close()
    return False

with open("список группы.txt", "r",encoding= 'UTF-8') as groupF:
    with open("список присутсвующих.txt","r",encoding= 'UTF-8') as presF:
        num = groupF.readline().strip()
        print(num)
        stars.append(num)
        points.append(num)
        presLine = ""
        while(not presLine.__contains__("Сортировать по имени:")):
            presLine = presF.readline()
        presLine = presF.readline()
        groupLine = groupF.readline()
        while(groupLine!=""):
            if(groupLine=="\n"):
                groupLine = groupF.readline().strip()
                print("\n"+groupLine)
                stars.append(groupLine)
                points.append(groupLine)
                groupLine = groupF.readline().strip()
            if(presLine.strip()!="" and checkPres(presLine)):
                presLine = presLine.strip()
                groupLine = groupLine.strip()
                if(groupLine.__contains__(presLine[5:].strip())):
                    stars.append("*")
                    points.append("+")
                    print(groupLine, "--", presLine)
                    presLine = presF.readline()
                    groupLine = groupF.readline()
                else:
                    print(groupLine," -- ", "empty")
                    stars.append(" ")
                    points.append("o")
                    groupLine = groupF.readline()
            else:
                presLine = presF.readline()

    presF.close()
groupF.close()
```

**Шаг 4.** Вывести *нераспознанные* записи ***"ФИО в списе присутсвующих"***   

```python
print("Нераспознанные:")
for i in range(otsut.__len__()):
    print(otsut[i])
    
```

**Шаг 6.** Вывести столбик из отметок **\*** для ручного копирования в электронную таблицу

```python
with open("список звёзд.txt",'w',encoding='UTF-8') as outF:
    for i in range(stars.__len__()):
        print(stars[i])
        outF.write(stars[i]+"\n")
    print("---------")
    outF.write("-----------------Конец------------------------")
outF.close()
```

**Шаг 7**. Вывести список из пар: отметка  **\[+\]** или **\[o\]** -- ФИО для ручного заполения ведомости в ***cabinet.ssau.ru***   

```python
with open("список ведомости.txt",'w',encoding='UTF-8') as outF:
    for i in range(points.__len__()):
        print(points[i])
        outF.write(points[i]+"\n")
    print("------------------------------")
    outF.write("------------------------------")
outF.close()
```
